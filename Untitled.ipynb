```python
import pandas as pd
from nltk.translate.meteor_score import meteor_score
import sacrebleu
import nltk

# Download required nltk resources
nltk.download('wordnet')
nltk.download('punkt')

# Load Excel
df = pd.read_excel("Evaluation_Set_With_ChatGPT_Translations_50.xlsx")
df = df.dropna(subset=["Ground_Truth", "ChatGPT_Translation"])

# Prepare lists
references = df["Ground_Truth"].str.strip().tolist()
hypotheses = df["ChatGPT_Translation"].str.strip().tolist()

# BLEU and TER
bleu = sacrebleu.corpus_bleu(hypotheses, [references]).score
ter = sacrebleu.corpus_ter(hypotheses, [references]).score

# METEOR (tokenized input)
meteor_scores = [
    meteor_score([ref.split()], hyp.split()) 
    for ref, hyp in zip(references, hypotheses)
]
avg_meteor = sum(meteor_scores) / len(meteor_scores)

# chrF++
chrf = sacrebleu.corpus_chrf(hypotheses, [references], word_order=2).score  # word_order=2 means chrF++

# Results
print(f"BLEU: {bleu:.2f}")
print(f"TER: {ter:.2f}")
print(f"METEOR: {avg_meteor:.4f}")
print(f"chrF++: {chrf:.2f}")
```
