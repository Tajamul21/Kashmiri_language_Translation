```python

# Load the dataset (without headers)
file_new_data = "/home/shubham/Kashmiri_language_Translation/new_data.xlsx"
df_new = pd.read_excel(file_new_data, header=None)

# Define column names
column_names = [
    "English_Word", "Part_of_Speech", "English_Sentence",
    "Kashmiri_Word", "Kashmiri_Translation", "Kashmiri_Sentence",
    "Urdu_Translation", "Urdu_Word"
]

# Assign column names based on structure
df_new.columns = [
    "English_Word", "English_Sentence", "Kashmiri_Word",
    "Kashmiri_Sentence", "Urdu_Word", "Urdu_Translation"
]

# Add missing columns with "NaN"
df_new["Part_of_Speech"] = "NaN"
df_new["Kashmiri_Translation"] = "NaN"

# Reorder columns to match the required format
df_new = df_new[column_names]

# Drop columns that contain only NaN values
for col in ["Kashmiri_Translation", "Part_of_Speech"]:
    if df_new[col].isna().all() or (df_new[col] == "NaN").all():
        df_new.drop(columns=[col], inplace=True)

# Swap "Urdu_Word" and "Urdu_Translation" (Column Names & Data)
df_new = df_new.rename(columns={"Urdu_Word": "TEMP", "Urdu_Translation": "Urdu_Word"})
df_new = df_new.rename(columns={"TEMP": "Urdu_Translation"})
df_new[["Urdu_Word", "Urdu_Translation"]] = df_new[["Urdu_Translation", "Urdu_Word"]]

# Save the cleaned dataset
output_file = "/home/shubham/Kashmiri_language_Translation/new_data_cleaned.xlsx"
df_new.to_excel(output_file, index=False)
```

```python

# Load dataset
file_4000_new = "/home/shubham/Kashmiri_language_Translation/4000_new.xlsx"
df_4000 = pd.read_excel(file_4000_new, header=None)

# Rename columns to match 5000_Parallel Corpora
df_4000.columns = [
    "English_Word", "Part_of_Speech", "English_Sentence",
    "Kashmiri_Word", "Kashmiri_Translation", "Kashmiri_Sentence",
    "Urdu_Translation", "Urdu_Word"
]

# Convert all columns to string type before filling NaN
df_4000 = df_4000.astype(str)

# Fill missing values with "NaN"
df_4000.replace("nan", "NaN", inplace=True)

# Swap only the data of "English_Sentence" and "Kashmiri_Translation"
df_4000[["English_Sentence", "Kashmiri_Translation"]] = df_4000[["Kashmiri_Translation", "English_Sentence"]].values

# Swap only the data of "Kashmiri_Translation" and "Urdu_Translation"
df_4000[["Kashmiri_Translation", "Urdu_Translation"]] = df_4000[["Urdu_Translation", "Kashmiri_Translation"]].values

# Swap only the data of "Kashmiri_Word" and "Urdu_Word"
df_4000[["Kashmiri_Word", "Urdu_Word"]] = df_4000[["Urdu_Word", "Kashmiri_Word"]].values

# Check if non-essential columns contain only NaN, then remove them
columns_to_keep = ["English_Sentence", "Kashmiri_Translation"]
columns_to_check = [col for col in df_4000.columns if col not in columns_to_keep]

for col in columns_to_check:
    if df_4000[col].isna().all() or (df_4000[col] == "NaN").all():
        df_4000.drop(columns=[col], inplace=True)

# Save the cleaned dataset
output_file = "/home/shubham/Kashmiri_language_Translation/4000_new_cleaned.xlsx"
df_4000.to_excel(output_file, index=False)
```

```python
# Load the dataset (update the file path accordingly)
file_4000_new = "/home/shubham/Kashmiri_language_Translation/Kash Eng 150 sents.xlsx"
df_kash_eng_150 = pd.read_excel(file_4000_new, header=None)

# Define column names based on 5000_Parallel Corpora
column_names = [
    "English_Word", "Part_of_Speech", "English_Sentence",
    "Kashmiri_Word", "Kashmiri_Translation", "Kashmiri_Sentence",
    "Urdu_Translation", "Urdu_Word"
]

# Assign column names while keeping the detected structure intact
df_kash_eng_150.columns = column_names[:df_kash_eng_150.shape[1]]

# Add missing columns and fill with "NaN"
for col in column_names:
    if col not in df_kash_eng_150.columns:
        df_kash_eng_150[col] = "NaN"

# Ensure all NaN values are stored as strings to avoid dtype issues
df_kash_eng_150 = df_kash_eng_150.astype(str)

# Swap only the data of "English_Sentence" and "Kashmiri_Translation"
df_kash_eng_150[["English_Sentence", "Kashmiri_Translation"]] = df_kash_eng_150[["Kashmiri_Translation", "English_Sentence"]].values

# Swap only the data of "Kashmiri_Translation" and "Urdu_Translation"
df_kash_eng_150[["Kashmiri_Translation", "Urdu_Translation"]] = df_kash_eng_150[["Urdu_Translation", "Kashmiri_Translation"]].values

# Check if non-essential columns contain only NaN, then remove them
columns_to_keep = ["English_Sentence", "Kashmiri_Translation"]
columns_to_check = [col for col in df_kash_eng_150.columns if col not in columns_to_keep]

for col in columns_to_check:
    if df_kash_eng_150[col].isna().all() or (df_kash_eng_150[col] == "NaN").all():
        df_kash_eng_150.drop(columns=[col], inplace=True)

# Remove rows where all values are "NaN"
df_kash_eng_150 = df_kash_eng_150[~(df_kash_eng_150 == "NaN").all(axis=1)]

# Save the modified dataset
output_file = "/home/shubham/Kashmiri_language_Translation/Kash_Eng_150_Cleaned.xlsx"
df_kash_eng_150.to_excel(output_file, index=False)
```

```python
import pandas as pd
from nltk.translate.meteor_score import meteor_score
import sacrebleu
import nltk
nltk.download('wordnet')
nltk.download('punkt')

# Load Excel
df = pd.read_excel("Evaluation_Set_With_ChatGPT_Translations_new.xlsx")
df = df.dropna(subset=["Ground_Truth", "ChatGPT_Translation"])

# Prepare lists
references = df["Ground_Truth"].str.strip().tolist()
hypotheses = df["ChatGPT_Translation"].str.strip().tolist()

# BLEU and TER
bleu = sacrebleu.corpus_bleu(hypotheses, [references]).score
ter = sacrebleu.corpus_ter(hypotheses, [references]).score


# METEOR (now with tokenized input)
meteor_scores = [
    meteor_score([ref.split()], hyp.split()) 
    for ref, hyp in zip(references, hypotheses)
]
avg_meteor = sum(meteor_scores) / len(meteor_scores)

# Results
print(f"BLEU: {bleu:.2f}")
print(f"TER: {ter:.2f}")
print(f"METEOR: {avg_meteor:.4f}")
```

```python
import pandas as pd

# List of all cleaned dataset files
file_paths = [
    "/DATA/Shubham/Projects/Kashmiri_language_Translation/5000_Parallel_Corpora_Cleaned.xlsx",
    "/DATA/Shubham/Projects/Kashmiri_language_Translation/4000_new_cleaned.xlsx", 
    "/DATA/Shubham/Projects/Kashmiri_language_Translation/500_new_cleaned.xlsx", 
    "/DATA/Shubham/Projects/Kashmiri_language_Translation/Kash_Eng_150_Cleaned.xlsx",
    "/DATA/Shubham/Projects/Kashmiri_language_Translation/new_data_cleaned.xlsx"
]

# Load and combine datasets
df_list = [pd.read_excel(file, dtype=str) for file in file_paths]
df_combined = pd.concat(df_list, ignore_index=True)

# Fill missing Kashmiri_Translation using Kashmiri_Sentence
df_combined['Kashmiri_Translation'] = df_combined.apply(
    lambda row: row['Kashmiri_Sentence'] if pd.isna(row['Kashmiri_Translation']) or str(row['Kashmiri_Translation']).strip() == '' else row['Kashmiri_Translation'],
    axis=1
)

# Filter: Keep only rows where English_Sentence and Kashmiri_Translation are both non-empty
df_filtered = df_combined[
    df_combined['English_Sentence'].notna() & df_combined['English_Sentence'].str.strip().ne('') &
    df_combined['Kashmiri_Translation'].notna() & df_combined['Kashmiri_Translation'].str.strip().ne('')
]

# Save the filtered dataset
df_filtered.to_excel("/DATA/Shubham/Projects/Kashmiri_language_Translation/Final_English_Kashmiri_Corpora.xlsx", index=False)
```
