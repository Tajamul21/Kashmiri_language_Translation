```python
# Load the dataset (update the file path accordingly)
file_4000_new = "/home/shubham/Desktop/Kashmiri_language_Translation/Kash Eng 150 sents.xlsx"
df_kash_eng_150 = pd.read_excel(file_4000_new, header=None)

# Define column names based on 5000_Parallel Corpora
column_names = [
    "English_Word", "Part_of_Speech", "English_Sentence",
    "Kashmiri_Word", "Kashmiri_Translation", "Kashmiri_Sentence",
    "Urdu_Translation", "Urdu_Word"
]

# Assign column names while keeping the detected structure intact
df_kash_eng_150.columns = column_names[:df_kash_eng_150.shape[1]]

# Add missing columns and fill with "NaN"
for col in column_names:
    if col not in df_kash_eng_150.columns:
        df_kash_eng_150[col] = "NaN"

# Ensure all NaN values are stored as strings to avoid dtype issues
df_kash_eng_150 = df_kash_eng_150.astype(str)

# Swap only the data of "English_Sentence" and "Kashmiri_Translation"
df_kash_eng_150[["English_Sentence", "Kashmiri_Translation"]] = df_kash_eng_150[["Kashmiri_Translation", "English_Sentence"]].values

# Swap only the data of "Kashmiri_Translation" and "Urdu_Translation"
df_kash_eng_150[["Kashmiri_Translation", "Urdu_Translation"]] = df_kash_eng_150[["Urdu_Translation", "Kashmiri_Translation"]].values

# Check if non-essential columns contain only NaN, then remove them
columns_to_keep = ["English_Sentence", "Kashmiri_Translation"]
columns_to_check = [col for col in df_kash_eng_150.columns if col not in columns_to_keep]

for col in columns_to_check:
    if df_kash_eng_150[col].isna().all() or (df_kash_eng_150[col] == "NaN").all():
        df_kash_eng_150.drop(columns=[col], inplace=True)

# Remove rows where all values are "NaN"
df_kash_eng_150 = df_kash_eng_150[~(df_kash_eng_150 == "NaN").all(axis=1)]

# Save the modified dataset
output_file = "/home/shubham/Desktop/Kashmiri_language_Translation/Kash_Eng_150_Cleaned.xlsx"
df_kash_eng_150.to_excel(output_file, index=False)
```

```python
import pandas as pd

# List of all cleaned dataset files
file_paths = [
    "5000_Parallel_Corpora_Cleaned.xlsx",
    "4000_new_cleaned.xlsx", 
    "500_new_cleaned.xlsx", 
    "Kash_Eng_150_Cleaned.xlsx",
    "new_data_cleaned.xlsx"
]

# Load all datasets into a list
df_list = [pd.read_excel(file, dtype=str) for file in file_paths]

# Concatenate all datasets
df_combined = pd.concat(df_list, ignore_index=True)

# Drop rows with missing values
df_filtered = df_combined.dropna(subset=["English_Sentence", "Kashmiri_Translation"])

# Sample 500 sentence pairs
sampled_data = df_filtered[["English_Sentence", "Kashmiri_Translation"]].sample(n=200, random_state=42)

# Rename columns
sampled_data = sampled_data.rename(columns={
    "English_Sentence": "English",
    "Kashmiri_Translation": "Ground_Truth"
})

# Add empty columns for future translations
sampled_data["ChatGPT_Translation"] = ""
# sampled_data["KashmiriGPT_Translation"] = ""

# Save the final structured file
output_file = "Random_200_Sentences_Structure.xlsx"
sampled_data.to_excel(output_file, index=False)

print(f"Structured file saved successfully as {output_file}")
```

```python
import pandas as pd
from nltk.translate.meteor_score import meteor_score
import sacrebleu
import nltk
nltk.download('wordnet')
nltk.download('punkt')

# Load Excel
df = pd.read_excel("Evaluation_Set_With_ChatGPT_Translations200.xlsx")
df = df.dropna(subset=["Ground_Truth", "ChatGPT_Translation"])

# Prepare lists
references = df["Ground_Truth"].str.strip().tolist()
hypotheses = df["ChatGPT_Translation"].str.strip().tolist()

# BLEU and TER
bleu = sacrebleu.corpus_bleu(hypotheses, [references]).score
ter = sacrebleu.corpus_ter(hypotheses, [references]).score


# METEOR (now with tokenized input)
meteor_scores = [
    meteor_score([ref.split()], hyp.split()) 
    for ref, hyp in zip(references, hypotheses)
]
avg_meteor = sum(meteor_scores) / len(meteor_scores)

# Results
print(f"BLEU: {bleu:.2f}")
print(f"TER: {ter:.2f}")
print(f"METEOR: {avg_meteor:.4f}")
```
